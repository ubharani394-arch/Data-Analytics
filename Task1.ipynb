```python
# ==========================================
# TASK 1: DATA IMMERSION & WRANGLING
# ==========================================

import pandas as pd
import numpy as np

# ------------------------------------------
# STEP 1: CREATE SAMPLE DATASET
# ------------------------------------------

data = {
    "Order_ID": [101, 102, 103, 103, 104],
    "Customer": ["John", "Sarah", "Mike", "Mike", "Emma"],
    "Age": [25, np.nan, 30, 30, 28],
    "City": ["New York", "Chicago", "New York", "New York", np.nan],
    "Sales": [500, 700, 500, 500, 900]
}

df = pd.DataFrame(data)

print("\n========== ORIGINAL DATA ==========")
print(df)

# ------------------------------------------
# STEP 2: DATASET INFORMATION
# ------------------------------------------

print("\n========== DATASET INFO ==========")
print(df.info())

print("\n========== DATA TYPES ==========")
print(df.dtypes)

# ------------------------------------------
# STEP 3: CHECK MISSING VALUES
# ------------------------------------------

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

# ------------------------------------------
# STEP 4: CHECK DUPLICATES
# ------------------------------------------

print("\n========== DUPLICATE ROWS ==========")
print(df.duplicated().sum())

# ------------------------------------------
# STEP 5: HANDLE MISSING VALUES
# ------------------------------------------

# Fill missing Age with mean age
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Fill missing City with 'Unknown'
df["City"] = df["City"].fillna("Unknown")

print("\n========== AFTER FILLING MISSING VALUES ==========")
print(df)

# ------------------------------------------
# STEP 6: REMOVE DUPLICATES
# ------------------------------------------

df = df.drop_duplicates()

print("\n========== AFTER REMOVING DUPLICATES ==========")
print(df)

# ------------------------------------------
# STEP 7: FEATURE ENGINEERING
# ------------------------------------------

def age_group(age):
    if age < 30:
        return "Young"
    elif age < 50:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)

print("\n========== AFTER FEATURE ENGINEERING ==========")
print(df)

# ------------------------------------------
# STEP 8: BASIC DATA ANALYSIS
# ------------------------------------------

print("\n========== SUMMARY STATISTICS ==========")
print(df.describe())

print("\n========== TOTAL SALES ==========")
print(df["Sales"].sum())

print("\n========== AVERAGE SALES ==========")
print(df["Sales"].mean())

print("\n========== SALES BY CITY ==========")
print(df.groupby("City")["Sales"].sum())

# ------------------------------------------
# STEP 9: SAVE CLEANED DATASET
# ------------------------------------------

df.to_csv("cleaned_sales_data.csv", index=False)

print("\n✅ Cleaned dataset saved as 'cleaned_sales_data.csv'")

# ------------------------------------------
# FINAL DATASET
# ------------------------------------------

print("\n========== FINAL CLEANED DATASET ==========")
print(df)
```
